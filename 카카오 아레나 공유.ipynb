### 데이터 분석 EDA (Exploratory Data Analysis: 탐색적 데이터 분석)

기초 데이터를 활용하여 필요한 정보를 탐색적으로 분석해 나가는 데이터 전처리 기법 중에 하나

- 링크: [EDA](https://arena.kakao.com/forum/topics/191)


- 고찰점
  - 파일 기반의 빅데이터(json 등)를 DB를 거치지 않고 분석 수행. 매우 빠름
    - join, group by, merge 등 가능
  - 바로바로 시각적으로 활용 가능
    - 도표, 그래프, 테이블 등등
  - 단, 경험이 있는 사람이 없고, 구성원들 모두 숙련도가 낮음
  - 데이터 분석 내용과 문제 해결을 연결시키기가 어려웠음
    - 단순 데이터 분포 확인 했는데, 그 다음에는..?

### 추천 모델 알고리즘 아이디어 논의

- 문제 플레이리스트 데이터 유형 별로 추천 알고리즘 분기 필요
  - 태그가 없는 경우
  - 노래가 없는 경우
  - 둘 다 없고, 제목만 있는 경우


- 도메인 지식 및 경험 (기획적 사고방식)
  - 플레이리스트의 성향적 유형 별로 문제 풀이 전략 수립 필요
  - 좋아요 수가 많으면 대중적인 곡..?


- 멜 스펙트로그램 음원 데이터 라벨링 작업한 뒤 학습시키기
- 협업 필터링 (티브로드)

### 문제풀이 시도

- 플레이리스트 제목 활용
  1. 한글 형태소 분석 라이브러리를 활용해 제목에서 한글 명사 추출
  2. 추출한 명사를 태그로 활용
  3. 그 태그로 train 플레이리스트를 검색한 뒤, 해당 플레이리스트의 노래를 추천
  4. 해당 태그가 train 플레이리스트에 없을 경우 Word2Vec 라이브러리를 활용해 유사 단어로 진행


- 멜 스펙트로그램 음원 데이터 라벨링 및 학습
