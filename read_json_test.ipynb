```python
val_df = pd.read_json('res/val.json', typ='frame')
```

### case 별 갯수 카운팅

전체

```python
len(val_df)
```

tag가 있는 경우

```python
val_df[val_df['tags'].apply(len).gt(0)]
```

song이 있는 경우

```python
val_df[val_df['songs'].apply(len).ne(0)]
```

태그와 노래

```python
val_df[val_df['tags'].apply(len).eq(0) & val_df['songs'].apply(len).eq(0)]
```

```python
val_df[val_df['plylst_title'] == '']
```

※ all, any 사용 예

```python
L = ['a', 'b', 'c', 'd', 'e']
items = 'a', 'b', 'c' # type tuple

# if 'a' in L and 'b' in L and 'c' in L):
if all(i in L for i in items): # and, generator comprehension
    print("모두 있음 !")
elif any(i in L for i in items): # or
    print("하나라도 있음 !")
else:
    print("False...")
```

```python
tag_series = train_df['tags']
```

```python
# list comprehension
target_tags = [tags for tags in tag_series if any(t in tags for t in ['비', '분위기', '컨트리', '서정적인'])]
```
